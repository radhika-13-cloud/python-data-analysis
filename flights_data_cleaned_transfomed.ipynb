```python
#convert columns to clean snake_case

df.columns = (df.columns.str.strip().str.lower().str.replace(" ", "_"))
```

```python
#check duplictaes

df.duplicated().sum()
```

```python
#remove duplictaes
df=df.drop_duplicates()
```

```python
#recheck if duplicates is removed or not
df.duplicated().sum()
```

```python
#fill missing airlines

df=df.copy()
df['airline']=df['airline'].fillna('unknown')
```

```python
#fill missing cancellation flags

df['cancelled']=df['cancelled'].fillna(0)
```

```python
#fill delays

delay_cols = ['departure_delay', 'arrival_delay']
for col in delay_cols:
    df[col] = df[col].fillna(0)
```

```python
#fill tail number

df['tail_number']=df['tail_number'].fillna('NA')
```

```python
#fill remarks text

df['remarks']=df['remarks'].fillna('NO REMARK')
```

```python
# convert date to datetime

df['flight_date']=pd.to_datetime(df['flight_date'],errors='coerce')
```

```python
# convert numeric  columns

num_cols=['departure_delay','arrival_delay','distance','cancelled']
df[num_cols]=df[num_cols].apply(pd.to_numeric, errors='coerce')
```

```python
#Strip spaces and fix case

df['airline']=df['airline'].str.strip().str.upper()
df['origin']=df['origin'].astype(str).str.strip().str.upper()
df['destination']=df['destination'].astype(str).str.strip().str.upper()
df['tail_number']=df['tail_number'].astype(str).str.strip().str.upper()
```

```python
#Standardize names
df['airline'] = df['airline'].replace({
    'AIR INDIA': 'AIR INDIA',
    'INDIGO': 'INDIGO',
    'SOUTHWEST': 'SOUTHWEST',
    '  SOUTHWEST  ': 'SOUTHWEST'
})
```

```python
#Remove distance ≤ 0

df=df[df['distance']>0]
```

```python
#Remove extremely wrong delays

df=df[(df['arrival_delay']>=-60) & (df['departure_delay']>=-60)]
```

```python
# creating new column total_delay

df['total_delay'] = df['arrival_delay'] + df['departure_delay']
```

```python
#flight status

df['status']=np.where(df['cancelled']==1, 'CANCELLED','COMPLETED')
```

```python
df.head()
```

```python
#delay category

def delay_bucket(x):
    if x <=0:
        return 'on time'
    elif x<=60:
        return 'slight delay'
    else:
        return 'heavy delay'
    

df['delay_category']=df['total_delay'].apply(delay_bucket)
```

```python
#export cleaned dataset

df.to_csv("flights_cleaned_final.csv", index=False)
```
